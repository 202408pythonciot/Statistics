## 散布図の作成
pandasでデータを読み込み，matplotlibで散布図を作成します。

```python
blue_coffee_shop_ice = pd.read_csv("./blue_coffee_shop_1_ice.csv")
blue_coffee_shop_hot = pd.read_csv("./blue_coffee_shop_2_hot.csv")
```

```python
plt.scatter(blue_coffee_shop_ice['最高気温'], blue_coffee_shop_ice['売上個数'],label='土日', color='#88F')
plt.title('ブルーコーヒー店（アイスコーヒー）')
plt.xlabel('最高気温')
plt.ylabel('売上個数')
plt.xlim([26, 36])
plt.ylim([0, 600])
plt.grid()
plt.show()
```

### 平日と土日を色分け
`query`メソッドを使うと，行の内容によるフィルタリングができます。"曜日"列の内容が土or日か，それ以外でプロットを分けて散布図を作成します。

```python
plt.scatter(blue_coffee_shop_ice.query('曜日 == "土" or 曜日 == "日"')['最高気温'], blue_coffee_shop_ice.query('曜日 == "土" or 曜日 == "日"')['売上個数'], label='土日', color='#F88')
plt.scatter(blue_coffee_shop_ice.query('not (曜日 == "土" or 曜日 == "日")')['最高気温'], blue_coffee_shop_ice.query('not (曜日 == "土" or 曜日 == "日")')['売上個数'], label='平日', color='#88F')
plt.title('ブルーコーヒー店（アイスコーヒー）')
plt.xlabel('最高気温')
plt.ylabel('売上個数')
plt.xlim([26, 36])
plt.ylim([0, 600])
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

### ホットコーヒーの散布図

```python
plt.scatter(blue_coffee_shop_hot['最高気温'], blue_coffee_shop_hot['売上個数'],label='土日', color='#88F')
plt.title('ブルーコーヒー店（ホットコーヒー）')
plt.xlabel('最高気温')
plt.ylabel('売上個数')
plt.xlim([26, 36])
plt.ylim([0, 600])
plt.grid()
plt.show()
```
